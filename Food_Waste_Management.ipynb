```python
import pandas as pd
from database import create_all_tables, insert_dataframe, clear_table

# Create tables and clear existing data
create_all_tables()
clear_table("providers")
clear_table("receivers")
clear_table("food_listings")
clear_table("claims")

# Load CSV files
df_providers = pd.read_csv("data/providers_data.csv")
df_receivers = pd.read_csv("data/receivers_data.csv")
df_listings = pd.read_csv("data/food_listings_data.csv")
df_claims = pd.read_csv("data/claims_data.csv")

# Fix date format in listings (convert to ISO yyyy-mm-dd)
df_listings["Expiry_Date"] = pd.to_datetime(df_listings["Expiry_Date"], errors="coerce").dt.strftime("%Y-%m-%d")

# Insert data into DB
insert_dataframe(df_providers, "providers")
insert_dataframe(df_receivers, "receivers")
insert_dataframe(df_listings, "food_listings")
insert_dataframe(df_claims, "claims")
```
